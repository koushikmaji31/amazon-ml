```python
# train.ipynb
import pandas as pd
from ocr import extract_text
from multiprocessing import Pool
import os

# Load your image paths
# Assuming you have a DataFrame 'df' with a column 'image_path'
df = pd.read_csv('tarin.csv')  # Replace with your actual CSV file
image_paths = df['image_path'].tolist()

# Define a function to process each image
def process_image(image_path):
    try:
        return extract_text(image_path)
    except Exception as e:
        print(f"Error processing {image_path}: {e}")
        return ""

# Use multiprocessing to speed up the process
with Pool(8) as pool:  # Adjust the number of processes based on your system
    results = pool.map(process_image, image_paths)

# Store the results
df['extracted_text'] = results
df.to_csv('extracted_texts.csv', index=False)
```
